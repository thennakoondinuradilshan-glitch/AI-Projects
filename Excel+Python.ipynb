```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("superstore.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['sales']=df['sales'].fillna(df['sales'].mean())
#Fill missing values in 'sales' column with the average sales value/Impute numerical null value with mean
```

```python
df['product_name']=df['product_name'].fillna('unknown')
## Fill missing values in 'product_name' column with 'unknown'/impute categorical value with anything you want
```

```python
df.duplicated().sum()
```

```python
#remove duplicate values
df.drop_duplicates(inplace=True)
```

```python
# remove
df.rename(columns={'product_name':'product'},inplace=True)
```

```python
df.columns
```

```python
#changing Data types
df['quantity']=df['quantity'].astype(int)
```

```python
#pivot table using groupby()
df.groupby("region")["sales"].sum().sort_values(ascending=False)
```

```python
#profit by catogory
df.groupby("category")["profit"].sum().sort_values(ascending=False)
```

```python
#Top 10 products
df.groupby('product')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
#visualization
#sales by region
df.groupby('region')['sales'].sum().plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
# profit by catogory
df.groupby('category')['profit'].sum().plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
df.groupby('category')['quantity'].sum().plot(kind='pie')
```

```python
df.groupby('category')['quantity'].sum()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# KPI Analysis
#total sales
df['TotalAmount'].sum()
```

```python
#total Orders
df['OrderID'].count()
```

```python
# Average sales
df['TotalAmount'].mean()
```

```python
#maximum Quantity
df['Quantity'].max()
```

```python
#REVENUE BY STATE
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
#Top 5 customers
df.groupby('CustomerName')['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
#orders byPayment more
df.groupby('PaymentMethod')['OrderID'].count().sort_values(ascending=False)
```

```python
#visualization
df.groupby('State')['TotalAmount'].sum().plot(kind='bar')
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=True).plot(kind='bar')
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```
